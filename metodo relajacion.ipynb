```python
from numpy import loadtxt,array, arange, linspace, sin, cos,exp,log, arcsin,zeros
from pylab import plot, show, savefig, ylim, xlim, grid, xlabel, ylabel
import matplotlib.pyplot as plt
import math
from math import tanh, cosh
```

```python
x0=1
x=[]
n=100
x.append(x0)
for i in range(n):
    x0=2-exp(-x0)
    x.append(x0)
plot(x)
print(x0)
show()
```

```python
x0=1/2
x=[]
n=100
x.append(x0)
for i in range(n):
    x0=exp(1-(x0*x0))
    x.append(x0)
plot(x)
print(x0)
show()
```

```python
x0=1/2
x=[]
n=100
x.append(x0)
for i in range(n):
    x0=(1-math.log(x0))**(1/2)
    x.append(x0)
plot(x)
print(x0)
show()
```

```python
x0=1/2
x=[]
n=100
x.append(x0)
for i in range(n):
    x0=(x0*x0)+sin(2*x0)
    x.append(x0)
plot(x)
print(x0)
show()
```

```python
x0=1/2
x=[]
n=100
x.append(x0)
for i in range(n):
    x0=(1/2)*arcsin(x0-(x0*x0))
    x.append(x0)
plot(x)
print(x0)
show()
```

```python
#ejemplo 6.3
Tmax=2.0
points = 1000
accuracy = 1e-2

y=[]
temp=linspace(0.01,Tmax,points)

for T in temp:
    m1=1.0
    error=1.0
    
    while error>accuracy:
        m1,m2 = tanh(m1/T),m1
        error=abs((m1-m2)/(1-T*cosh(m2/T)**2))
    y.append(m1)

plot(temp,y)
ylim(-0.1,1.1)
xlabel("Temperatura")
ylabel("Magnetizacion")
show()
```
